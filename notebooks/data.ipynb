# Data Inventory

**Purpose:** quick overview of everything in `GPS_DATA_DIR` — what files exist, and for
the `gps_data*.csv` track files specifically: which months they cover, how many months,
how many distinct days, and a preview of each.

```python
import sys
sys.path.append("..")

from pathlib import Path

import pandas as pd

from gps_lib import config
```

```python
data_dir = Path(config.DATA_DIR)
print(f"Data directory: {data_dir}\n")

for p in sorted(data_dir.iterdir()):
    if p.is_dir():
        n_files = sum(1 for _ in p.iterdir())
        print(f"[dir]  {p.name}/  ({n_files} files)")
    else:
        size_mb = p.stat().st_size / 1024 / 1024
        print(f"       {p.name}  ({size_mb:.2f} MB)")
```

```python
gps_files = sorted(data_dir.glob("gps_data*.csv"))
print(f"Found {len(gps_files)} gps_data CSV file(s):")
for f in gps_files:
    print(" -", f.name)
```

```python
# Month/day coverage for each gps_data file
summary_rows = []

for f in gps_files:
    df = pd.read_csv(f, usecols=["get_time"])
    times = pd.to_datetime(df["get_time"], errors="coerce")
    months = sorted(times.dt.to_period("M").dropna().unique().astype(str))
    summary_rows.append({
        "file": f.name,
        "rows": len(df),
        "n_months": len(months),
        "months": ", ".join(months),
        "n_days": times.dt.date.nunique(),
        "min_time": times.min(),
        "max_time": times.max(),
    })

summary_df = pd.DataFrame(summary_rows)
summary_df
```

```python
# Preview each gps_data file
for f in gps_files:
    print(f"\n=== {f.name} ===")
    display(pd.read_csv(f).head(3))
```

```python
# Preview the other data files (tracker list, zone metadata, etc.)
other_files = sorted(set(data_dir.glob("*.csv")) - set(gps_files))
for f in other_files:
    print(f"\n=== {f.name} ===")
    display(pd.read_csv(f).head())
```
