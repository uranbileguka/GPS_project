# Fetch Tracker List

**Purpose:** authenticate against the Navixy API, pull the current tracker/vehicle list, classify each as dump/loader/other, and save it to `data/tracker_list.csv`.

This is a metadata refresh — run it occasionally (e.g. when vehicles are added/retired), not on every data pull. For the actual GPS point history, see `fetch_gps_tracks.ipynb`.

```python
import sys
sys.path.append("../..")

from gps_lib import navixy_api, classify, io_utils
```

```python
hash_token = navixy_api.authenticate()
print("Authenticated.")
```

```python
tracker_list_df = navixy_api.get_tracker_list(hash_token)
tracker_list_df.shape
```

```python
tracker_list_df = classify.classify_technic_type(tracker_list_df)
tracker_list_df.groupby("technic_type")["label"].count()
```

```python
out_path = io_utils.save_csv(tracker_list_df, "tracker_list.csv")
print("Saved:", out_path)
```
